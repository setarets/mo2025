## Задание 2

Прочитать данные `https://raw.githubusercontent.com/yakushinav/omo/main/data/playstore.xlsx`.

Для столбца `Last_Updated` установить формат `дата_и_время`.

````python
import pandas as pd
````

````python
df = pd.read_excel('https://raw.githubusercontent.com/yakushinav/omo/main/data/playstore.xlsx').head()

df['Last_Updated'] = pd.to_datetime(df['Last_Updated'])
df['Last_Updated'] = df['Last_Updated'].dt.strftime('%Y-%m-%d')
print(df.head())
````
